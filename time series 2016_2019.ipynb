```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
Sweden_file_path = "D:\DE_Jacobs University\Spring 2020\Advance Project I\Sweden\Sweden_energy.csv"
Sweden_df = pd.read_csv(Sweden_file_path)
```

```python
# new data frame with split dates column
new_df = Sweden_df["Time (CET)"].str.split("-", n=1, expand=True)
Sweden_df["DATE Part 1"]= new_df[0]
Sweden_df["CET_TimeStamp"] = new_df[1]
```

```python
Sweden_df.head()
```

```python
Sweden_df = Sweden_df.drop(columns=['Time (CET)', 'DATE Part 1','Solar Generation Actual', 'Solar Generation forecast'])
```

```python
Sweden_df = Sweden_df[['CET_TimeStamp', 'Actual Total Load', 'Day Ahead Price', 'Wind Onshore Gen Actual', 
                       'Wind Onshore generation forecast']]
```

```python
Sweden_df.head()
```

```python
Sweden_df.rename(columns={'Actual Total Load': 'Actual_Total_Load'}, inplace=True)
Sweden_df.rename(columns={'Day Ahead Price': 'Day_Ahead_Price'}, inplace=True)
Sweden_df.rename(columns={'Wind Onshore Gen Actual': 'Wind_Onshore_Gen_Actual'}, inplace=True)
Sweden_df.rename(columns={'Wind Onshore generation forecast': 'Wind_Onshore_generation_forecast'}, inplace=True)
```
